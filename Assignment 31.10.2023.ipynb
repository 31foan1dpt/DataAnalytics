```python
import numpy as np
import pandas as pd
```

```python
#read-in data as pandaFrame
df = pd.read_csv('data/US_Accidents_March23.csv')
#convert to numpy-Array in order to execute ops 
np_array = df.to_numpy()
print(np_array)
```

```python
#filter data with 2 conditions
sourceArray = np_array[:,1]
severityArray = np_array[:,2]
temperatureArray = np_array[:,20]
filtered_data_severity = severityArray[np.argwhere(severityArray > 3)]
print(filtered_data_severity)
filtered_data_source = sourceArray[np.argwhere(sourceArray == 'Source1')]
print(filtered_data_source)
```

```python
#2 statistics
print(np.mean(severityArray))
print(np.median(temperatureArray))
```

```python
#calculate new column
df['source_severity'] = df['Source'].astype(str) + df['Severity'].astype(str)
print(df['source_severity'])
```

```python
#change naming
df.rename(columns={'source_severity':'severity_source'})
```

```python
#change order of columns
cols = list(df.columns)
cols[-2], cols[-1] = cols[-1], cols[-2]
df = df[cols]
print(df)
```

```python
#delete column severity_source
df = df.drop('source_severity', axis=1)
print(df)
```

```python
#use str-Accesor --> replace ID (A-1 --> 1)
df['ID'] = df['ID'].str.replace('A-','')
print(df)
```

```python
df.to_csv('data/edited_file.csv')
```
