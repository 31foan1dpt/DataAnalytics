```python
#Now that we gained experience with the dataset and the python-bibs we can do some further datacleaning and conversion for the features we want to use in the modeling part
#import bibs
import pandas as pd
```

```python
#read in data 
df_accidents = pd.read_csv('data/US_Accidents_March23.csv', index_col = 'ID')
```

```python
#output data
df_accidents
```

```python
#get datatypes 
df_accidents.dtypes
#features that we want to convert: Start_time/End_Time, Description, Street, City, County, State, Zipcode, Country, Timezone
```

```python
df_accidents.isna().sum()
```

```python
#show numbers of nan-values
df_accidents.isna().sum()
```

```python
for column in df_accidents.columns:
    nan_count = df_accidents[column].isna().sum()
    print(f"Column '{column}' has {nan_count} NaN values.")
```

```python
#Drop columns that are not needed like 'Source', 'Wind_Chill(F)' --> not significant
#'End_Lat' and 'End_Lng' are also not significant, because we focus on the start points (\"cause\" areas) of the accidents

df_accidents.drop('Source', axis=1, inplace=True)
df_accidents.drop('Wind_Chill(F)', axis=1, inplace=True)
df_accidents.drop('End_Lat', axis=1, inplace=True)
df_accidents.drop('End_Lng', axis=1, inplace=True)
```

```python
#Drop also 'Percipitation(in)' because auf too many NaN values
df_accidents.drop('Precipitation(in)', axis=1, inplace=True)
```

```python
#Drop Weather_Timestamp: The most timestamps are recorded within 10-15 minutes after the accident started. Despite that, there are over 120k NaN values in the column
df_accidents.drop('Weather_Timestamp', axis=1, inplace=True)
```

```python
#Drop all rows with at least one NaN value
df_accidents_cleaned = df_accidents.dropna()
print("Count of rows after cleaning: ", df_accidents_cleaned.shape[0])
```

```python
#display amount fo
df_accidents_cleaned.isna().sum()
```

```python
df_accidents_cleaned.to_csv('data/edited_file.csv')
```
