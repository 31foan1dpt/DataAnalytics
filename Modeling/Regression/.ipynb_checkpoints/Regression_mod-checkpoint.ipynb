```python
# Change names of some columns for the equation (because it leads to errors otherwise)

accidents = accidents.rename(columns={
    'Temperature(F)': 'Temperature',
    'Humidity(%)': 'Humidity',
    'Visibility(mi)': 'Visibility',
    'Pressure(in)': 'Pressure',
    'Wind_Speed(mph)': 'Wind_Speed',
    'Distance(mi)': 'Distance',
    'State Name': 'StateName',
})
```

```python
severity_basic = ols("Severity ~ Weather_Condition", accidents).fit()
print(severity_basic.summary())
```

```python
# BASIC: Syntax für die Regressionsgleichung: Zielwert ~ STATE

severity_basic = ols("Severity ~ State", accidents).fit()
print(severity_basic.summary())
```

```python
# FULL: Syntax für die Regressionsgleichung: Zielwert ~ State + Strassengegebenheiten + Wetterkonditionen
#900secs bis hier hin

severity_basic = ols("Severity ~ + Start_Lat + Start_Lng + Distance + State + Timezone + Temperature + Humidity + Pressure + Visibility + Wind_Speed + Weather_Condition + Amenity + Bump + Crossing + Give_Way + Junction + No_Exit + Railway + Roundabout + Station + Stop + Traffic_Calming + Traffic_Signal + Turning_Loop +  Year + Month + Day + Hour + Duration", accidents).fit()
print(severity_basic.summary())
```
